# Regex in Python and Pattern Matching 

```python
import re
```

```python
# . matches all the patterns
text="Hello my name is Sian Rodrigues D-098"
newlist=re.findall('.',text)
print(newlist)
```

```python
# * to match with 0 or more 
n1="Hello Worlddd thiss is Sian"    
print(re.findall('ld*',n1))
# Matches with hello and and world but he--l--l 
```

```python
# + matches with one 
print(re.findall('ld+',n1))
```

```python
# Exclusively for 0 or One repititions immediately after preceding it
# ? zero or one expression immediatley after precedding it  
print(re.findall('l?',n1))
# 0--> Expressions are also possible 
```

```python
# Match any character in the Square curly braces 
print(re.findall('S[ia]n','Name is Sian Rods'))
print(re.findall('\-[0-9]',n1))
```

```python
# all characters except in S --> [^S]
#  Any RE matches A|B
# Matches the end --> $ 
#  Used to drop a special meaning of the character --> '\'
print(re.findall('S[^ia]$',"Sian Sia"))
```

```python
print(re.findall('be[lt]s','Nice belts there'))
```
