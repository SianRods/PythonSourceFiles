## Using Pie-Charts to represent small data in the quantity of 3-5 at max

```python
from matplotlib import pyplot as plt 
import numpy as np
```

#### Pie-charts are mostly used to denote the proportions of a quantity with respect to other quantities 
1. `plt.pie(numerical_list,labels_for_data,explode_values,wedgeprops,color,autopct='%1.1f%%')`
2. Numerical List denotes the actual values to be compared for plotting pie-charts
3. labels contains the string values representing the numerical list in order 
4. explode values takes in the list of values specifying the offset of a pie w.r.t centre 
5. wedgeprops specifies the properties of the wedge 
6. color takes a hex , string values of colours to be used on the plot
7. `autopct=` can be used to format and add percentage values on pie 

```python
slices = [59219,55466,47544,36443,35917]
labels_lst=['JavaScript','HTML/CSS','SQL','Python','Java']  #Has to be mentioned in the order as the numerical values 

explode_values=[0.1,0,0,0,0]
colors_lst=['Blue','Green','Orange','Pink','Purple']
```

```python
plt.style.use("fivethirtyeight")
plt.pie(slices,labels=labels_lst,explode=explode_values,wedgeprops={'edgecolor':'black'},autopct='%1.1f%%')
plt.title("Programming Language Pie-Charts")
# I f we don't specify any colors the by-default colors of the choosen style will be used 
```
