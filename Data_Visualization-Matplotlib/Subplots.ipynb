# Plotting Different plots using the Object Oriented Method

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Importance of Subplots
 `subplots()` in Matplotlib is considered an object-oriented approach to plotting. In this method, instead of relying on the state-based functions (like `plt.plot()`), you work directly with objects like `Figure` and `Axes`, which gives you more control over how your plots are rendered and customized.

### Why `subplots()` is Object-Oriented:

- **`Figure` and `Axes` objects**: When you call `subplots()`, it returns two key objects: 
  - **Figure**: This is the overall container that holds all the plots (or subplots). Think of it as the canvas on which the individual plots are drawn.
  - **Axes**: These represent the individual plots themselves. Each `Axes` object corresponds to a single plot (or subplot), and you can manipulate these objects directly to customize titles, labels, or the content of the plot.

### Key Points of the Object-Oriented Approach:

1. **Direct interaction with Axes**: Instead of using high-level global commands like `plt.plot()`, you interact directly with the `Axes` objects. This allows you to modify each plot independently, such as adding titles, labels, and legends to individual subplots.

2. **Greater flexibility**: This approach is particularly useful when you're working with multiple plots or subplots. Since you have individual control over each `Axes` object, you can easily adjust each subplot’s appearance, layout, and content in a customized way.

3. **Explicit control**: The object-oriented method gives you explicit control over your figures and axes. For instance, if you're creating complex visualizations with multiple subplots, this approach makes it easy to manage and configure each part of the plot without confusion.

### Comparison with State-Based Approach:

In the state-based approach, you use global commands like `plt.plot()` to quickly generate plots. This method is straightforward for simple tasks but lacks the fine-grained control that the object-oriented approach provides. When you're dealing with complex visualizations or want to customize multiple plots, the object-oriented approach using `subplots()` is more scalable and robust. 

Overall, the object-oriented method is preferred for creating complex visualizations, especially when you need to manage multiple subplots or make specific customizations to each plot.

## Parameters in the Subplots 
The `subplots()` function in Matplotlib is used to create a figure with multiple subplots, and it provides several parameters to control the layout and appearance of these subplots. Here’s an explanation of the key parameters:

### 1. **`nrows` and `ncols`**
   - These parameters specify the number of rows and columns in the subplot grid. 
   - **`nrows`** defines how many rows of subplots there will be.
   - **`ncols`** defines how many columns of subplots there will be.
   - For example, if `nrows=2` and `ncols=3`, you will have a grid of 2 rows and 3 columns, giving you a total of 6 subplots.

### 2. **`figsize`**
   - This parameter defines the size of the entire figure (the canvas), and it is given as a tuple of width and height in inches (e.g., `(10, 5)`).
   - By adjusting this, you can control how big or small the overall figure appears. This does not change the number of subplots, but rather the overall dimensions of the figure.

### 3. **`sharex` and `sharey`**
   - These parameters control whether the subplots share the same x-axis or y-axis scales.
   - **`sharex=True`** makes all subplots in a row share the same x-axis labels and limits.
   - **`sharey=True`** makes all subplots in a column share the same y-axis labels and limits.
   - Sharing axes is useful when you want all subplots to be compared on the same scale without repeating axis labels.

### 4. **`squeeze`**
   - This parameter determines whether the output of `subplots()` should be simplified (or "squeezed") when there is only one row or column of subplots.
   - If **`squeeze=True`** (the default), and you have only one row or one column, the `Axes` object is returned as a 1D array or even a single object (for a 1x1 grid).
   - If **`squeeze=False`**, the `Axes` are always returned as a 2D array, even if there is only one row or column.

### 5. **`subplot_kw`**
   - This parameter allows you to pass a dictionary of keyword arguments to customize each individual subplot (`Axes` objects).
   - For example, you can specify properties like the projection type (e.g., 3D) or the aspect ratio of each subplot.
   - It’s useful for applying specific settings to all subplots at once.

### 6. **`gridspec_kw`**
   - This is another dictionary of keyword arguments, but it applies to the overall grid layout of the subplots (not the individual subplots themselves).
   - It allows you to control aspects like the spacing between subplots (e.g., `wspace`, `hspace`), margins, and subplot size ratios.

### 7. **`fig` and `axs` (returned objects)**
   - After calling `subplots()`, two objects are returned: `fig` (the Figure) and `axs` (the Axes objects).
   - **`fig`**: Represents the entire figure or canvas that contains all subplots.
   - **`axs`**: Represents the individual subplot(s), returned either as a single `Axes` object or an array of `Axes` objects, depending on how many subplots are created. You can manipulate these objects directly to adjust each subplot (e.g., set labels, titles, plot content).

### 8. **`constrained_layout`**
   - This parameter helps to automatically adjust the spacing between subplots to prevent overlap of axis labels, titles, and other elements.
   - **`constrained_layout=True`** ensures that the figure layout automatically adjusts to fit the content better without overlapping elements.

### 9. **`tight_layout` (alternative)**
   - Similar to `constrained_layout`, but often used as a method after subplots are created. It adjusts the layout to make sure that subplots are not too close together or overlapping.



These parameters give you full control over how multiple plots are arranged and displayed in a single figure.

```python
# Extracting the Data 
data = pd.read_csv('subplots_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
# Plotting the Data using the plt Method Together
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**What if we want to create seperate plot for all developrs and to achieve this we can use the subplot method of plt** 

```python
fig,ax=plt.subplots() 
# Here the ax represents the axis and fig represents the figure 
ax.plot(ages,py_salaries,label='Python')
ax.plot(ages,dev_salaries,label='Developers')
ax.plot(ages,js_salaries,label='JavaScript')
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')
ax.legend();
```

**Here Note that we have to use `set_xlabel`,`set_ylabel`,`set_title` when we use the axis term using the subplots and default the for subplots the `nrows=1`,`ncols=1`**

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
# When wed specify nrows=2 and ncols=1 we create a grid of plots containing 2 Rows and 1 Cols 
print(f"The ax contains the following information : {ax}") 
print(f"The Figure contains the following information : {fig}")
```

```python
fig,ax=plt.subplots(nrows=3,ncols=4)
# When wed specify nrows=2 and ncols=1 we create a grid of plots containing 2 Rows and 1 Cols 
print(f"The ax contains the following information : {ax}") 
print(f"The Figure contains the following information : {fig}")
```

```python
fig , ax=plt.subplots(nrows=2,ncols=1)
# We know that ax will ret
ax1,ax2=ax
# We will use the ax1 instance 
ax1.plot(ages,py_salaries,label='Python',color='red',marker='+')
ax1.plot(ages,js_salaries,label='JavaScript',color='green')
ax2.plot(ages,dev_salaries,label='Developers')
ax1.set_ylabel("Median Salary(in USD)")
ax2.set_ylabel("Median Salary(in USD)")
ax1.legend()
ax2.legend()
ax2.set_xlabel("Ages")
```

**Now in order to share the same 'x-axis' marking**

```python
fig , ax=plt.subplots(nrows=2,ncols=1,sharex=True)
# We know that ax will ret
ax1,ax2=ax
# We will use the ax1 instance 
ax1.plot(ages,py_salaries,label='Python',color='red')
ax1.plot(ages,js_salaries,label='JavaScript',color='green')
ax2.plot(ages,dev_salaries,label='Developers')
ax1.set_ylabel("Median Salary(in USD)")
ax2.set_ylabel("Median Salary(in USD)")
ax1.legend()
ax2.legend()
ax1.set_title("Plot Comparisons using Subplots")
ax2.set_xlabel("Ages")
```
