3. Given the Iris dataset, identify and print rows where sepal length (cm) is
 an outlier using the IQR method.

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
q1 = iris_df['sepal length (cm)'].quantile(0.25)
q3 = iris_df['sepal length (cm)'].quantile(0.75)
iqr = q3-q1
outliers = iris_df[(iris_df['sepal length (cm)'] < (q1 - 1.5 * iqr)) | (iris_df['sepal length (cm)']> (q3+1.5*iqr))]
outliers.head()
```
