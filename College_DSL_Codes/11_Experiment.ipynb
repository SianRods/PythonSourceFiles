 7.In the Boston Housing dataset, remove rows where the CHAS column

```python
boston_df = pd.read_csv("D:\Python Source Files\College_DSL_Codes\HousingData.csv")
boston_df.head()
```

```python
boston_df = boston_df.dropna(subset=['CHAS'])
boston_df['CHAS'].unique()
```

8.In the Boston Housing dataset, detect and remove outliers in the TAX
 column using the IQR method.

```python
q1 = boston_df['TAX'].quantile(0.25)
q3 = boston_df['TAX'].quantile(0.75)
iqr = q3-q1
outliers = boston_df[(boston_df['TAX']< q1 - 1.5 * iqr) | (boston_df['TAX']> q1 + 1.5 * iqr)]
boston_df = boston_df.drop(outliers.index)
```

9.In the Boston Housing dataset, replace missing values in the DIS

12. In the Wine Quality dataset, identify and drop outliers in the pH column
 using Z- scores.

```python
from scipy import stats
z_scores = stats.zscore(wine_df['pH']) 
wine_df['pH_zscore'] = z_scores
outliers = wine_df[wine_df['pH_zscore'].abs() > 3] 
wine_df.drop(outliers.index, inplace=True)
```
