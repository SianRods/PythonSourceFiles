```python
import numpy as np 
```

Various types of arithemtic operations can be performed on *Similar sized Array* or *By Broadcasting on Dissimilar Types*

```python
arr1 = np.array([[1,2,3],
				 [4,5,6]])
arr2 = np.array([[9,8,7],
				 [1,2,3]])
```

**Addition of Two Arrays**

```python
arr_add=arr1+arr2
print(arr_add)
# Element Wise Addition is being performed here 
arr_sub=arr1-arr2
print(arr_sub)
```

**Scalar Multiplication and Not Matrix Multiplication**

```python
# Note that multiplying two matrices elements --> executes it element wise unlike Mathematical 
# Multiplication
print(arr1*arr1)

# Similarly Dividing it by a Matrix 
print(arr1/arr1) # Division takes Element wise 

print(arr1**2)
```

## Basic Indexing and Slicing in case of Multi-Dimensional Arrays 

Array Indexing and Slicing is important topic of Numpy Arrays as Subset Selection and other data analytical approaches can be executed using Array Indexing and Slicing 

**An important first distinction from Pythonâ€™s built-in lists is that array slices are views on the original array.This means that the data is not copied, and any modifications to the view will be reflected in the source array.**
So any changes made in the view arrays will be directly reflected in the main array 

```python
sample_arr=np.arange(10)
print(sample_arr)

# Slicing out certaing portion 
sel_ar=sample_arr[3:7]
print(sel_ar)
# Similar convention to python is Followed here where the last index is excluded [first_index:last_index]

sel_ar[:]=12
print(sample_arr)
# sel_ar[:]=12 Selects all the Elements in the given array
# Such above slicing is also called as the 'Bare' Slicing 

sample_arr[1:4]=5.54 
print(sample_arr)# Automatic Conversion to 	Type int64


```

**In order to avoid overwriting the main array we can instead use the copy function to store it differently**

```python
sample_copy=sample_arr[3:7].copy()
sample_copy[2]=1234
print(sample_copy)
print(sample_arr)

# Nohing gets relfected , Infact the copy method enables us to create a new array
```

### Slicing in 2 Dimensional Arrays 


```python
print(arr1)
print(arr1[0]) # Printing only the first row
print(arr1[1])

print(arr1[:1,1])  # Selecting till the 1st Row 'and' the 2nd column element
# For printing 6 as just an element --> Element Extraction 
print(arr1[1,2]) 

# For printing 6 as an array--> Slicing
print(arr1[1:,2])  # Selecting from 1st row till end and the 2nd element 
```

**Difference between `arr1[1,2]=6` && `arr1[1:,2]=[6]` can be realised using the type function on them**

```python
print(type(arr1[1:,2]))
print(type(arr1[1,2]))
```

### Boolean Indexing and Arrays 
*Boolean indexing is a powerful technique in NumPy that allows you to select elements from an array based on a boolean condition. This condition is essentially a boolean array (an array containing only True and False values) that acts as a mask, filtering the original array*

**One of the most common uses involves filtering data based on conditions**



```python
# Lets first Consider a Normal Textbook example :-
newar=np.random.randn(3,3)
# Generating a random 3*3 TWO Dimensional Array
print(newar)

# Lets Consider a random string array which is correlated with this 3*3 Numeric array
# If i want to generate some conditions depending on String Array amd directly apply it on the
# Numeric array same can be done using the boolean array

name_data=np.array(['Bob', 'Joe', 'Will']) 

# Condition Based Generation of a boolean array
print(name_data[name_data=="Bob"]) # Gives array containing the name 'Bob'
print(name_data=='Bob')

print("The dtype of the given array is : ",(name_data=="Bob").dtype)

# Now i can apply the masking on the main array 

cond=name_data=="Bob"

print(newar[cond])  # Print only the first row 
# Notice that the Boolean array must have the same dimension as the main numeric array 
# In order to apply the condition 
# If it doesnt have the dim we can make so by slicing /indexing the nominal array in this case
```

***Applying the condition in case of irregular shape by Slicing or Indexing***

```python
# Consider the same array 
print(newar)
# Let the Conditional array be now say binary with different Dimensions 
binarr=np.array([1,1,0,0,1,0,1])

# Notice that the dimensions in the newar is 3*3 and in binarr is 7*1
# 7--> 3 to apply masking on the newar
print("Displaying the first two rows depending on the high conditions using slicing  : ")
newar[binarr[:3]==1]

# Similarly a negation condiition can also be applied on the same 
print("Applying indexing as well as Slicing ")
print(newar[~(binarr[:3]==1),:])
```

*Boolean selection will not fail if the boolean array is not the correct
 length, so I recommend care when using this feature.*

**Selecting data from an array by boolean indexing always creates a copy of the data,
 even if the returned array is unchanged**

```python
# Multiple Conditions can also be combined --> &(and ) or | (or)

# Setting the values in the Main Array Using Boolean Conditions also Works perfectly fine :-
newar[newar<0]=35
newar

newar[newar<0.35573773]=908
newar

# Such operations are more easier with pandas
```

```python
big_arr= np.arange(9).reshape(3,3)

# When we pass and an Integer array in the indexing of an Numpy Array is is Called as Fancy indexing  

print(big_arr)
big_arr[[1,]]  # Example of Fancy Indexing (Passing an Integer Array)

# Using the negative indexing 
big_arr[[-1,-2]]

```

**Passing Multiple Index Arrays**

```python
# When Multiple Index arrays are passed the corressponding tuple indices elements are choosen 

print(big_arr)
big_arr[[0,2],[1,2]]

# The Corresponding tuple arrays are : --> () (0,1)  (2,2) in format (row,column)
```
