### Conditional Logic as Array Operations 

**`where()` function**

```python
# Excuting condtional based approach based on pure python is not efficient 
# A typical use of where in data analysis is to produce a new array
#  of values based on another array

arr1=np.random.rand(3,3)
print("Array Number 1 : ",arr1)

# format of the where funciton 
#  np.where(base_condition, if_true_use_this, else_use_this )
# Both Scalar and arrays can be combined and used in case of where 

print("Filtering the array using the condition :\n",np.where(arr1<0.5,0.0,1))

# Note that such methods can be used to filter out necessary data and analyse it quickly 

arr2 = np.random.rand(3,3)
print("Array Number 2 : \n",arr2)

# Combining Arrays based on Condition 
# comb about points where value in 1st is less that 0.5

print("Result : \n",np.where(arr1<0.5,arr2,arr1))
```

**Mathematical & Statistical `Methods()`**

```python
# Top Level Method --> Method present in the main numpy module 
# Array Method --> Methods present in the Array class of NumPy
#  Columns (axis=1)--> All the data Across the Columns 
# Rows (axis =0) --> All the data down the rows 
# If the axis is not specified all the elements would be considered 
statsarr=np.random.rand(4,3)
print(statsarr)

# Using the Top Level Method -->
print("The mean of the entire matrix is : \n",np.mean(statsarr))

# using the method of the array class 
print("Using the array class method : \n",statsarr.mean())
print("Standard Deviation : \n",statsarr.std())

# Computing the Cumulative Sums Starting from 0
# cumpord --> Starting from 1
# Retruns an array of same lenght

arr1d=np.arange(10)
print("The Cumulative Sum of the given array is : ",arr1d.cumsum()) 
print("The Cummulative Product of the given array is ",arr1d.cumprod())
# Although if cumprod() starts from 1 if element in an array is 0 --> cumprod=0 
```

```python
# cumsum() and cumprod() incase of 2-D Array 

print("Cum Sum across the columns is : \n",statsarr.cumsum(axis=1))
print("Cum product down the rows is : \n",statsarr.cumprod(axis=0))
```

### Methods for Boolean Arrays : 

```python
# 1 --> True  0--> False 
# When we use sum() on boolean array it calculates the Number of True(1's)
boolar=np.random.rand(10000)
# Consider a random array of size 10000 
#  In this case counting 10000 will be very easier using the sum function
print("The Number of Positive Elements  is : ",(boolar>0.023).sum())
print("Random Condition Check : ",(boolar<0.324).sum()	)
```

**Any() and All() Methods**

```python
# The Any () methods checks if any of the following more than 1 or 1 value is true 
space=np.array([1,2,3,45,1,2,1])
print("Checking for any Ones :",space.any())
print("Checking for any Zeros :",~space.any())

alltrue=np.array([1,1,1,1,1])
print("If all are true values : ",alltrue.all())
```

## Sorting Values : -

```python
# # In order to Compute certain Statistical values we need to arrange given data in order 
# #  ex). Quantiles ,etc
# #  sort() --> Method

unsortar=np.random.rand(6)
print(unsortar)
# Sorting the array 
unsortar.sort() # Overides the original array and converts it to sorted format 
print(unsortar)
```

**Sorting in case of Multi-Dimensional Arrays**

```python
# Generating a random array : -
multidim = np.random.rand(5,5)
print(multidim)

# While using sort function on mutlidimension always use the keyword of axis (0/1)
multidim.sort(1)
print("Sorted array across the columns : \n",multidim)
multidim.sort(0)
print("Sorted array down the rows  : \n",multidim)
```
